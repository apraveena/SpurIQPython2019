```python
#board
board = ["-","-","-",
         "-","-","-",
         "-","-","-"]
game_still_going = True
curr_player = "X"
winner = "None"

#display board
def display_board():
        print(board[0] + " | " + board[1] + " | " + board[2])
        print(board[3] + " | " + board[4] + " | " + board[5])
        print(board[6] + " | " + board[7] + " | " + board[8])



def play_game():
    global winner
    #display initial board
    display_board()
    while game_still_going:
        handle_turn()
        check_for_winner()
    if(winner == "X" or winner == "O"):
        print (winner + " won!!")
    elif (winner == "None"):
        print("It's a tie!!")

def handle_turn():
    position = input("Enter your position from 1 to 9: ")
    position = int(position) - 1
    board[position] = curr_player
    display_board()
    flip_player()
    #handle_game_still_going()

def check_for_winner():
    check_if_win()
    check_if_tie()

def check_if_win():
    global winner
    row_winner = check_rows()
    col_winner = check_columns()
    diag_winner = check_diagonal()
    if (row_winner != ""):
        winner = row_winner
    elif (col_winner != ""):
        winner = col_winner
    elif (diag_winner != ""):
        winner = diag_winner
    else:
        winner = "None"
    

def check_rows():
    #set global variable
    global game_still_going
    #check if anyone won
    row_1 = board[0] == board[1] == board[2] != "-"
    row_2 = board[3] == board[4] == board[5] != "-"
    row_3 = board[6] == board[7] == board[8] != "-"
    
    #return the winner
    if(row_1 or row_2 or row_3):
        game_still_going = False
        if row_1:
            return board[0]
        elif row_2:
            return board[3]
        elif row_2:
            return board[6]
    return

def check_columns():
    #set global variable
    global game_still_going
    #check if anyone won
    col_1 = board[0] == board[3] == board[6] != "-"
    col_2 = board[1] == board[4] == board[7] != "-"
    col_3 = board[2] == board[5] == board[8] != "-"
    
    #return the winner
    if(col_1 or col_2 or col_3):
        game_still_going = False
        if col_1:
            return board[0]
        elif col_2:
            return board[1]
        elif col_3:
            return board[2]
    return

def check_diagonal():
    #set global variable
    global game_still_going
    #check if anyone won
    diag_1 = board[0] == board[4] == board[8] != "-"
    diag_2 = board[2] == board[4] == board[6] != "-"
    
    #return the winner
    if(diag_1 or diag_2):
        game_still_going = False
        if diag_1:
            return board[0]
        elif diag_2:
            return board[2]
    return

def check_if_tie():
    return

def flip_player():
    global curr_player
    if(curr_player == "X"):
        curr_player = "O"
    else:
        curr_player = "X" 
        
    
play_game()


    
```
