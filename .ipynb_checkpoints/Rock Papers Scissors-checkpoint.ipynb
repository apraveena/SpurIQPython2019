```python
import random

def convertLetterToWord(argument):
    switcher = {
        "p": "Paper",
        "r": "Rock",
        "s": "Scissors",
    }
    return switcher.get(argument, "nothing")

moves = ["r", "p", "s"]
player_wins = ["pr", "sp", "rs"]

player_score = 0
computer_score = 0

while (True):
    player_move = input ("Enter your move as r, p or s. Press q to quit ")
    if (player_move == "q"):
        break
    if(player_move not in "prsq"):
        continue
    computer_move = random.choice(moves)
    print ("Your move: " + convertLetterToWord(player_move))
    print ("Computer move: " + convertLetterToWord(computer_move))
    result = player_move + computer_move
    if(player_move == computer_move):
        print ("It's a tie")
    elif (result in player_wins):
        print ("You win")
        player_score += 1
    else:
        print ("Computer wins")
        computer_score += 1
    print ("Player score: " + str(player_score))
    print ("Computer score: " + str(computer_score))
    

    
```
